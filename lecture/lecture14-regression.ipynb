# Examining the daaset with tips

```python
import seaborn as sns
tips = sns.load_dataset("tips")
```

```python
tips.head(5)
```

```python
chart = alt.Chart(tips).mark_point().encode(
    x='total_bill', 
    y='tip'
)
chart
```

```python
tips['percent'] = round(tips['tip'] / tips['total_bill'], 2)
```

```python
tips
```

```python
alt.Chart(tips).mark_point().encode(
    x='total_bill', 
    y='percent'
)
```

```python
model.fit(X = tips[['total_bill']], y = tips['percent'])
# slope
b_hat_sklearn = model.coef_
# y-axis intercept
a_hat_sklearn = model.intercept_

print("Slope =", b_hat_sklearn, "intercept =", a_hat_sklearn)
```

```python
chart = alt.Chart(tips).mark_point().encode(
    x='total_bill', 
    y='percent'
)

reg_line = chart + chart.transform_regression('total_bill', 'percent').mark_line(color='red')

alt.layer(chart, reg_line).configure_axis(
    labelFontSize=14, # change axes label font size
    titleFontSize=16  # change axes title font size
)
```

```python
model.predict(np.array(50).reshape(-1,1))[0]
```

# One-hot encoder

```python
tips.head()
```

```python
# importing one hot encoder
from sklearn.preprocessing import OneHotEncoder
# creating one hot encoder object
onehotencoder = OneHotEncoder()
```

```python
#reshape the 1-D array to 2-D as fit_transform expects 2-D and finally fit the object
X = onehotencoder.fit_transform(tips[['day']]).toarray()
X
```

```python
from sklearn.feature_extraction import DictVectorizer

days = tips[['day']].to_dict(orient='records')
#days

encoder = DictVectorizer(sparse=False)
days_df = pd.DataFrame(
    data = encoder.fit_transform(days),
    columns = encoder.feature_names_
)
days_df
```

```python
tips.shape
```

```python
tips_day = pd.concat([tips, days_df], ignore_index=False, axis=1)
tips_day
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
predictors = tips_day[['total_bill', 'day=Fri', 'day=Sat', 'day=Sun', 'day=Thur']]
predictors.head()
```

```python
outcome = tips['tip']
outcome.head()
```

```python
model.fit(predictors, outcome)
```

```python
model.coef_, model.intercept_
```

```python
model.predict(predictors)
```
